```python
import json
import requests

# Load secrets from credentials.json
url = 'https://api.foursquare.com/v2/venues/explore'
with open('.secrets/credentials.json') as f: #if it starts with a dot, then it's a hidden file/folder 
    params = json.load(f)
#the .secrets folder is gitignored
#credentials has our specific API key and 
```

```python
#these parameters are found in the documentation
params['v'] = '20180323' #v = version, input is a date - will still keep old version from that date 
params['ll'] = '47.608,-122.336', #lattitude and longititude required unless global searching 
params['query'] = 'records',
params['intent'] = 'checkin',
params['radius'] = 5000,
params['limit'] = 10
```

```python
response = requests.get(url=url, params=params)
data = json.loads(response.text)
```
